```python
import numpy as np
theta = np.random.rand(3,4)
```

```python
n_iter = 1
```

```python
y_transformed = np.zeros((y.size, y.max()+1))
y_transformed[np.arange(y.size),y] = 1
print(y)
```

```python
print(y_transformed)
```

```python
n_iter=3000
threshold = float("inf")
eta = 0.1 
theta_th = theta
for i in range(n_iter):
  sk = X.dot(theta.T)
  sk_1 = np.exp(sk)
  sk_1 = sk_1/sk_1.sum(axis=1, keepdims=True)
  sk_2 = np.log(sk_1)
  J = (-1/m)*np.sum(np.multiply(y_transformed, sk_2))
  print(J)
  gradient = (1/m)*((sk_1-y_transformed).T.dot(X))
  theta = theta - eta*gradient
  if(J<= threshold):
    theta_th = theta
    threshold = J
   
```

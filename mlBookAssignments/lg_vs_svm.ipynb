```python
from sklearn import preprocessing

def preprocess_cat_col(df, encoded_labels, numeric_col):
    le = preprocessing.LabelEncoder()
    columns = df.columns
    for col in columns:
        if(col not in numeric_col):
            unique_len = len(df[col].unique())
            if((unique_len<=3) or (col in encoded_labels)):
                df[col]=le.fit_transform(df[col])
            else:
                X = pd.get_dummies(df[col])
                X = X.drop(X.columns[0], axis=1)
                df[X.columns] = X
                df.drop(col, axis=1, inplace=True)
            
            
    return df
    
df = preprocess_cat_col(bank_add_full, ['month','day_of_week'], bank_add_full._get_numeric_data().columns)    










```

```python
from sklearn.model_selection import train_test_split
target = df['y'].to_numpy()
data  = df.drop('y',1).to_numpy()



```

```python
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.25, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression()
logisticRegr.fit(x_train, y_train)
```

```python
predictions = logisticRegr.predict(x_test)
```

```python
score = logisticRegr.score(x_test, y_test)
print(score)
```

```python
from sklearn import svm
from sklearn import metrics
clf = svm.SVC(kernel='linear') 
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
clf_poly = svm.SVC(kernel='poly') 
clf_poly.fit(x_train, y_train)
y_pred_poly = clf_poly.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_poly))
```
